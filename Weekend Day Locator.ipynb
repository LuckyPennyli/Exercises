```python
import pandas as pd
from datetime import datetime,date
df = pd.read_csv("StartDate.csv")
df
```

```python
s = df.apply(lambda x:pd.date_range(start=x['StartDate'],periods=x['AmortPeriods'], freq="M"), axis=1)
df1 = pd.DataFrame(s)
df1.set_axis(['AmortDates'], axis = 'columns', inplace=True)
df1
```

```python
df1.dtypes
```

```python
df1['AmortDates'] = df1['AmortDates'].astype(str)
df1
```

```python
df3 = df1['AmortDates'].str.split(',', expand=True)
df3
```

```python
df4 = df3.add_prefix('C').astype(str)
df4
```

```python
str_cols = df4.select_dtypes(['object']).columns
df4[str_cols] = df4[str_cols].replace(r'\D+', '', regex=True)
df4 = df4.drop(['C60', 'C61'], axis=1)
df4
```

```python
df4.dtypes
```

```python
def convert_fill(df4):
    return df4.stack().apply(pd.to_datetime, format='%Y%m%d', errors='coerce').dt.date.fillna(0).unstack()
convert_fill(df4)
```

```python
df5 = convert_fill(df4)
cols = df5.columns[0:60]
df5[cols] = df5[cols].apply(pd.to_datetime, errors='coerce')
df5.dtypes
```

```python
import datetime
def convert_fill(df5):
    return df5.stack().apply(date.weekday).unstack()
convert_fill(df5)
```
